# Swarm of 6 robots

```python
import pandas as pd
import numpy as np
import random
```

# Intial Map

```python
x = np.ones((10, 10))
x[:, :] = 0
x
```

## Random position of 6 robots

```python
res = []
for j in range(6): 
    res.append(random.randint(0, 99))
res
```

```python
robotsX = []
robotsY = []
for i in range(6):
    robotsX.append(res[i]//10)
    robotsY.append(res[i]%10)
print(robotsX)
robotsY
```

```python
lastPosX = robotsX.copy()
lastPosY = robotsY.copy()
```

## Robots in those random positions: 1 stands for first robot, 2 for second robot and so on

```python
for i in range(6):
    x[robotsX[i]][robotsY[i]] = i+1
x
```

```python
InitialMap = x.copy()
```

## Simple swarm algorithm

```python
def swar():
    for i in range(6):
        if (((lastPosX[i]+1) < 10) and (x[lastPosX[i]+1][lastPosY[i]] == 0.0)):
            robotsX.append(lastPosX[i]+1)
            robotsY.append(lastPosY[i])
        
        elif (((lastPosX[i]-1) > -1) and (x[lastPosX[i]-1][lastPosY[i]] == 0.0)):
            robotsX.append(lastPosX[i]-1)
            robotsY.append(lastPosY[i])
        
        elif (((lastPosY[i]+1) < 10) and (x[lastPosX[i]][lastPosY[i]+1] == 0.0)):
            robotsY.append(lastPosY[i]+1)
            robotsX.append(lastPosX[i])
        
        elif (((lastPosY[i]-1) > -1) and (x[lastPosX[i]][lastPosY[i]-1] == 0.0)):
            robotsY.append(lastPosY[i]-1)
            robotsX.append(lastPosX[i])
        
        else:
            robotsX.append(lastPosX[i])
            robotsY.append(lastPosY[i])

    #print(robotsX)
    #print(robotsY)
    for i in range(len(robotsX)):
        x[robotsX[i]][robotsY[i]] = (i%6)+1
    print(x)
    for i in range(6):
        lastPosX[i] = robotsX[(len(robotsX)-1+i)-5]
        lastPosY[i] = robotsY[(len(robotsY)-1+i)-5]
    #print(lastPosX)
    #print(lastPosY)
```

## Running the algorithm for 20 steps - Trajectory of the robot in each step

```python
for i in range(20):
    print("Map in", i ,"th run" )
    swar()
```
